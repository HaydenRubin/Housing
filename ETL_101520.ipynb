```python
import pandas as pd
```

```python
file_path_2019='raw_data/2019Housing.csv'
file_path_2018='raw_data/2018Housing.csv'
file_path_2017='raw_data/2017Housing.csv'
file_path_2016='raw_data/2016Housing.csv'
file_path_2015='raw_data/2015Housing.csv'
file_path_2014='raw_data/2014Housing.csv'
file_path_2013='raw_data/2013Housing.csv'
file_path_2012='raw_data/2012Housing.csv'
file_path_2011='raw_data/2011Housing.csv'
file_path_2010='raw_data/2010Housing.csv'
```

```python
#function to transform the raw data
def data_transf_15to19(file_path,year):
    origin_df=pd.read_csv(file_path)
    
    extract_df=origin_df[['GEO_ID','NAME','DP04_0002PE',
                          'DP04_0115PE','DP04_0124PE','DP04_0063E','DP04_0064E','DP04_0065E','DP04_0066E',
                         'DP04_0067E','DP04_0068E','DP04_0069E','DP04_0070E','DP04_0081E','DP04_0082E',
                        'DP04_0083E','DP04_0084E','DP04_0085E','DP04_0086E','DP04_0087E','DP04_0088E','DP04_0089E']]
    
    renamed_df=extract_df.rename(columns={'NAME':'state','DP04_0002PE':'occupied_perc',
                                'DP04_0115PE':'smocapi_35ormore_m_perc',
                                'DP04_0124PE':'smocapi_35more_nm_perc','DP04_0063E':'utilitygas',
                                'DP04_0064E':'bottled_tank','DP04_0065E':'electricity',
                                'DP04_0066E':'fueloil','DP04_0067E':'coal_coke',
                                'DP04_0068E':'wood','DP04_0069E':'solar','DP04_0070E':'other_fuel',
                                'DP04_0081E':'value_50k_less','DP04_0082E':'value_50000to99999',
                                'DP04_0083E':'value_100000to149999','DP04_0084E':'vlaue_150000to199999',
                                'DP04_0085E':'value_200000to299999','DP04_0086E':'value_300000to499999',
                                'DP04_0087E':'value_500000to999999','DP04_0088E':'value_1M_more',
                                'DP04_0089E':'house_median_value'})
    
    renamed_df['year']=year
    renamed_df['date']=f'{year}-01-01'
    
    
    sorted_df=renamed_df.sort_values(by=['GEO_ID'])
    
    #row_df=reindex.loc[['California','Colorado','Connecticut']]
    
    row_df=sorted_df.iloc[1:52,1:]
    
    
    return row_df
```

```python
final_2019=data_transf_15to19(file_path_2019,2019)
final_2018=data_transf_15to19(file_path_2018,2018)
final_2017=data_transf_15to19(file_path_2017,2017)
final_2016=data_transf_15to19(file_path_2016,2016)
final_2015=data_transf_15to19(file_path_2015,2015)
```

```python
def data_transf_10to14(file_path,year):
    origin_df=pd.read_csv(file_path)
    
    extract_df=origin_df[['GEO_ID','NAME','DP04_0002PE',
                          'DP04_0113PE','DP04_0122PE','DP04_0062E','DP04_0063E','DP04_0064E',
                          'DP04_0065E','DP04_0066E','DP04_0067E','DP04_0068E','DP04_0069E','DP04_0080E',
                          'DP04_0081E','DP04_0082E','DP04_0083E','DP04_0084E','DP04_0085E',
                          'DP04_0086E','DP04_0087E','DP04_0088E']]
    
    renamed_df=extract_df.rename(columns={'NAME':'state','DP04_0002PE':'occupied_perc',
                                'DP04_0113PE':'smocapi_35ormore_m_perc',
                                'DP04_0122PE':'smocapi_35more_nm_perc','DP04_0062E':'utilitygas',
                                'DP04_0063E':'bottled_tank','DP04_0064E':'electricity',
                                'DP04_0065E':'fueloil','DP04_0066E':'coal_coke',
                                'DP04_0067E':'wood','DP04_0068E':'solar','DP04_0069E':'other_fuel',
                                'DP04_0080E':'value_50k_less','DP04_0081E':'value_50000to99999',
                                'DP04_0082E':'value_100000to149999','DP04_0083E':'vlaue_150000to199999',
                                'DP04_0084E':'value_200000to299999','DP04_0085E':'value_300000to499999',
                                'DP04_0086E':'value_500000to999999','DP04_0087E':'value_1M_more','DP04_0088E':'house_median_value'})
    
    renamed_df['year']=year
    renamed_df['date']=f'{year}-01-01'
    sorted_df=renamed_df.sort_values(by=['GEO_ID'])
    
    #row_df=reindex.loc[['California','Colorado','Connecticut']]
    
    row_df=sorted_df.iloc[1:52,1:]
    
    
    return row_df
```

```python
final_2014=data_transf_10to14(file_path_2014,2014)
final_2013=data_transf_10to14(file_path_2013,2013)
final_2012=data_transf_10to14(file_path_2012,2012)
final_2011=data_transf_10to14(file_path_2011,2011)
final_2010=data_transf_10to14(file_path_2010,2010)
```

```python
merged_df=pd.concat([final_2019,final_2018,final_2017,final_2016,final_2015,final_2014,final_2013,final_2012,final_2011,final_2010])
final_df=merged_df.reset_index(drop=True)

state_code_df=pd.read_csv("raw_data/states-code.csv")
state_codes=state_code_df.rename(columns={'State':'state','Abbrev':'abbrev','Code':'code'})

left_join=pd.merge(final_df,state_codes,on='state',how='left')
left_join['primary_key']=left_join['code']+'_'+left_join['year'].astype(str)
```

```python
from sqlalchemy import create_engine

# Imports the methods needed to abstract classes into tables
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

# Allow us to declare column types
from sqlalchemy import Column, Integer, String, Float

from sqlalchemy.orm import Session,sessionmaker
```

```python
# create your class
class USHOUSING(Base):
    __tablename__ = 'ushousing_10to19'
    __table_args__ = {'extend_existing': True}
    state = Column(String)
    occupied_perc = Column(Float)
    smocapi_35ormore_m_perc = Column(Float)
    smocapi_35more_nm_perc = Column(Float)
    utilitygas = Column(Integer)
    bottled_tank = Column(Integer)
    electricity = Column(Integer)
    fueloil = Column(Integer)
    coal_coke = Column(Integer)
    wood = Column(Integer)
    solar = Column(Integer)
    other_fuel = Column(Integer)
    value_50k_less = Column(Integer)
    value_50000to99999 = Column(Integer)
    value_100000to149999 = Column(Integer)
    vlaue_150000to199999 = Column(Integer)
    value_200000to299999 = Column(Integer)
    value_300000to499999 = Column(Integer)
    value_500000to999999 = Column(Integer)
    value_1M_more = Column(Integer)
    house_median_value = Column(Integer)
    year = Column(Integer)
    date = Column(String)
    abbrev = Column(String)
    code= Column(String(225))
    primary_key = Column (String(225),primary_key=True)
```

```python
database_path = "ushousing_db.sqlite"
engine = create_engine(f"sqlite:///{database_path}")
conn = engine.connect()
```

```python
# Create a "Metadata" Layer That Abstracts our SQL Database
# ----------------------------------
# Create (if not already in existence) the tables associated with our classes.
Base.metadata.create_all(engine)

# Use this to clear out the db
# ----------------------------------
#Base.metadata.drop_all(engine)
```

```python
Session = sessionmaker(bind=conn)
session = Session()
session.bulk_insert_mappings(USHOUSING, left_join.to_dict(orient="records"))
session.commit()
session.close()
```
