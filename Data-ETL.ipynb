```python
#Declar the file path of each year
file_path_2019='raw_data/USHousing-2019.csv'
file_path_2018='raw_data/USHousing-2018.csv'
file_path_2017='raw_data/USHousing-2017.csv'
file_path_2016='raw_data/USHousing-2016.csv'
file_path_2015='raw_data/USHousing-2015.csv'
file_path_2014='raw_data/USHousing-2014.csv'
file_path_2013='raw_data/USHousing-2013.csv'
file_path_2012='raw_data/USHousing-2012.csv'
file_path_2011='raw_data/USHousing-2011.csv'
file_path_2010='raw_data/USHousing-2010.csv'
```

```python
def data_transf(file_path,year):
    origin_df=pd.read_csv(file_path)
    
    extract_df=origin_df[['NAME','DP04_0001E','DP04_0002E','DP04_0002PE','DP04_0007E','DP04_0008E','DP04_0009E',
                       'DP04_0010E','DP04_0011E','DP04_0012E','DP04_0013E','DP04_0014E','DP04_0015E',
                       'DP04_0039E','DP04_0040E','DP04_0041E','DP04_0042E','DP04_0043E','DP04_0044E',
                       'DP04_0080E','DP04_0081E','DP04_0082E','DP04_0083E','DP04_0084E','DP04_0085E','DP04_0086E',
                       'DP04_0087E','DP04_0088E','DP04_0089E','DP04_0110E','DP04_0111E','DP04_0112E',
                       'DP04_0113E','DP04_0114E','DP04_0115E','DP04_0116E','DP04_0117E','DP04_0118E','DP04_0119E',
                       'DP04_0120E','DP04_0121E','DP04_0122E','DP04_0123E','DP04_0124E','DP04_0125E']]
    
    renamed_df=extract_df.rename(columns={'DP04_0001E':'Total Housing Units','DP04_0002E':'Occupied Units','DP04_0002PE':'Occupied Percent','DP04_0007E':'1-unit det',
                                'DP04_0008E':'1-unit att','DP04_0009E':'2 units','DP04_0010E':'3 or 4 units','DP04_0011E':'5 to 9 units','DP04_0012E':'10 to 19 units',
                                'DP04_0013E':'20 or more units','DP04_0014E':'Mobile home','DP04_0015E':'boat,RV,van,etc.','DP04_0039E':'No bedroom','DP04_0040E':'1 bedroom',
                                'DP04_0041E':'2 bedrooms','DP04_0042E':'3 bedrooms','DP04_0043E':'4 bedrooms','DP04_0044E':'5 or more bedrooms','DP04_0080E':'owner-occupied units',
                                'DP04_0081E':'Less than $50k','DP04_0082E':'$50,000 to $99,999','DP04_0083E':'$100,000 to $149,999','DP04_0084E':'$150,000 to $199,999',
                                'DP04_0085E':'$200,000 to $299,999','DP04_0086E':'$300,000 to $499,999','DP04_0087E':'$500,000 to $999,999','DP04_0088E':'$1M or more',
                                'DP04_0089E':'Median (dollars)','DP04_0110E':'units with mortgage with SMOCAPI','DP04_0111E':'SMOCAPI:less than 20 perc(m)','DP04_0112E':'SMOCAPI:20 to 24.9 perc(m)',
                                'DP04_0113E':'SMOCAPI:25 to 29.9 perc(m)','DP04_0114E':'SMOCAPI:30 to 34.9 perc(m)','DP04_0115E':'SMOCAPI:35 perc(m) or more','DP04_0116E':'SMOCAPI not computed(m)',
                                'DP04_0117E':'units without mortgage with SMOCAPI','DP04_0118E':'SMOCAPI:less than 10 perc(nm)','DP04_0119E':'SMOCAPI:10 to 14.9 perc(nm)',
                                'DP04_0120E':'SMOCAPI:15 to 19.9 perc(nm)','DP04_0121E':'SMOCAPI:20 to 24.9 perc(nm)','DP04_0122E':'SMOCAPI:25 to 29.9 perc(nm)',
                                'DP04_0123E':'SMOCAPI:30 to 34.9 perc(nm)','DP04_0124E':'SMOCAPI:35 perc(nm) or more','DP04_0125E':'SMOCAPI not computed(nm)'})
    
    renamed_df['Year']=year
    renamed_df['Date']=f'{year}-01-01'
    
    row_df=renamed_df.iloc[1:53,:]
    
    
    return row_df
```

```python
#cleanup/transform each dataset
final_2019=data_transf(file_path_2019,2019)
final_2018=data_transf(file_path_2019,2018)
final_2017=data_transf(file_path_2019,2017)
final_2016=data_transf(file_path_2019,2016)
final_2015=data_transf(file_path_2019,2015)
final_2014=data_transf(file_path_2019,2014)
final_2013=data_transf(file_path_2019,2013)
final_2012=data_transf(file_path_2019,2012)
final_2011=data_transf(file_path_2019,2011)
final_2010=data_transf(file_path_2019,2010)
```

```python
final_df=pd.concat([final_2019,final_2018,final_2017,final_2016,final_2015,final_2014,final_2013,final_2012,final_2011,final_2010])
final_df.shape
```

```python
reindex=final_df.reset_index(drop=True)
reindex.tail()
```

```python
reindex.to_csv('final_dataset.csv',index=False)
```
